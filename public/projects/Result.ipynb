```python
# =========================================
# IMPORT LIBRARIES
# =========================================

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# =========================================
# SHOW ALL ROWS AND COLUMNS
# =========================================

pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

```python
# =========================================
# LOAD CSV FILE
# =========================================

df = pd.read_csv(r"C:\DS\Result.csv")

# Remove extra spaces from column names
df.columns = df.columns.str.strip()

# Display Data
print(df)
```

```python
# =========================================
# BEAUTIFUL TABLE DISPLAY
# =========================================

display(
    df.style.set_properties(**{
        'text-align': 'center',
        'border': '1px solid black'
    }).set_table_styles([
        {
            'selector': 'th',
            'props': [
                ('background-color', 'red'),
                ('color', 'black'),
                ('text-align', 'center'),
                ('border', '1px solid black')
            ]
        }
    ])
)
```

```python
# =========================================
# SET NAME COLUMN AS INDEX
# =========================================

df.set_index('Name', inplace=True)

print(df)
```

```python
# =========================================
# BAR GRAPH
# =========================================

df.plot(kind='bar', figsize=(15,8))

plt.title("All Subjects Marks Graph")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.xticks(rotation=45)

plt.legend(title="Subjects")

plt.show()
```

```python
# =========================================
# LINE GRAPH
# =========================================

sns.set_style("whitegrid")

df.plot(kind='line', marker='o', figsize=(15,8))

plt.title("Students Marks in All Subjects")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.xticks(rotation=45)

plt.legend(title="Subjects")

plt.show()
```

```python
# =========================================
# AREA GRAPH
# =========================================

df.plot(kind='area', figsize=(15,8), alpha=0.5)

plt.title("All Subjects Area Graph")
plt.xlabel("Students")
plt.ylabel("Marks")

plt.xticks(rotation=45)

plt.legend(title="Subjects")

plt.show()
```

```python
# =========================================
# PIE CHART
# =========================================

# Reset index temporarily
df_reset = df.reset_index()

# Calculate Total Marks
df_reset['Total'] = df_reset.iloc[:, 1:].sum(axis=1)

plt.figure(figsize=(8,8))

plt.pie(
    df_reset['Total'],
    labels=df_reset['Name'],
    autopct='%1.1f%%'
)

plt.title("Students Total Marks Percentage")

plt.show()
```

```python
# =========================================
# HISTOGRAM
# =========================================

df.plot(kind='hist', figsize=(12,7), alpha=0.6)

plt.title("Histogram of Subjects Marks")
plt.xlabel("Marks")
plt.ylabel("Frequency")

plt.show()
```

```python
# =========================================
# BOX PLOT
# =========================================

plt.figure(figsize=(12,7))

sns.boxplot(data=df)

plt.title("Box Plot of Students Marks")

plt.show()
```

```python
# =========================================
# HEATMAP
# =========================================

plt.figure(figsize=(12,7))

sns.heatmap(df, annot=True, cmap='coolwarm')

plt.title("Students Marks Heatmap")

plt.show()
```

```python
# =========================================
# SCATTER PLOT
# =========================================

plt.figure(figsize=(10,6))

plt.scatter(df['Advance Java'], df['Flutter'])

plt.title("Advance Java vs Flutter")
plt.xlabel("Advance Java")
plt.ylabel("Flutter")

plt.show()
```

```python
# =========================================
# SUBJECT WISE AVERAGE MARKS
# =========================================

avg_marks = df.mean()

print(avg_marks)
```

```python
# =========================================
# AVERAGE MARKS BAR GRAPH
# =========================================

avg_marks.plot(kind='bar', figsize=(10,6))

plt.title("Average Marks of Subjects")
plt.xlabel("Subjects")
plt.ylabel("Average Marks")

plt.xticks(rotation=45)

plt.show()
```
